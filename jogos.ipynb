```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
from scipy import stats
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
import numpy as np 
from sklearn import datasets
from sklearn import linear_model
import statsmodels.api as sm 
```

```python
df = pd.read_excel('dados1.xlsx')
df = df.drop(["Unnamed: 0","Population","Online_Population",'Revenue'],axis = 1)
```

```python
y = df['Per_Capita_Revenue']
x = df.drop(['Per_Capita_Revenue'],axis=1)
```

```python
ln = LinearRegression()
```

```python
ln.fit(x,y)
print(ln.score(x,y),ln.coef_, ln.intercept_)
```

```python
yhat = ln.predict(x)
```
